```python
#import the required libraries

import matplotlib.pyplot as plt
import seaborn as sns
import keras
from keras.models import Sequential
from keras.layers import Dense, Conv2D , MaxPool2D , Flatten , Dropout 
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.regularizers import L2

from sklearn.metrics import classification_report,confusion_matrix

import tensorflow as tf

import cv2
import os

import numpy as np


labels = ['good', 'bad', 'nochange', 'invalid']
img_size = 224

#load the Dataset
data = [] 
     
#good
path = 'c:\\objects\\data\\train\\good'
#path = input("Source Train Good Images Folder") 


class_num = labels.index('good')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
         print(e)
#bad
path = 'c:\\objects\\data\\train\\bad'
#path = input("Source Train Bad Images Folder") 

class_num = labels.index('bad')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)
        
#nochange
path = 'c:\\objects\\data\\train\\nochange'
#path = input("Source Train NoChange Images Folder") 

class_num = labels.index('nochange')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)        
#train = np.array(data)

#invalid
path = 'c:\\objects\\data\\train\\invalid'
#path = input("Source Train Invalid Images Folder") 

class_num = labels.index('invalid')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)        
train = np.array(data)

data = [] 
     
#good
path = 'c:\\objects\\data\\test\\good'
#path = input("Source Test Good Images Folder") 

class_num = labels.index('good')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
         print(e)
#bad
path = 'c:\\objects\\data\\test\\bad'
#path = input("Source Test Bad Images Folder") 

class_num = labels.index('bad')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)
#nochnge
path = 'c:\\objects\\data\\test\\nochange'
#path = input("Source Test NoChange Images Folder") 

class_num = labels.index('nochange')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)     
        
#invalid
path = 'c:\\objects\\data\\test\\invalid'
#path = input("Source Test Invalid Images Folder") 

class_num = labels.index('invalid')
for img in os.listdir(path):
    try:
        img_arr = cv2.imread(os.path.join(path, img))[...,::-1] #convert BGR to RGB format
        resized_arr = cv2.resize(img_arr, (img_size, img_size)) # Reshaping images to preferred size
        data.append([resized_arr, class_num])
    except Exception as e:
        print(e)    
        
val = np.array(data)
```

```python
# Now we can easily fetch our train and validation data.

l = []
l.append("good")
l.append("bad")
l.append("nochange")
l.append("invalid")



x_train = []
y_train = []
x_val = []
y_val = []
#training data
for feature, label in train:
  x_train.append(feature)
  y_train.append(label)
    
#validation/test data
for feature, label in val:
  x_val.append(feature)
  y_val.append(label)
  

# Normalize the data
x_train = np.array(x_train) / 255
x_val = np.array(x_val) / 255

x_train.reshape(-1, img_size, img_size, 1)
y_train = np.array(y_train)

x_val.reshape(-1, img_size, img_size, 1)
y_val = np.array(y_val)
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape = (224, 224, 3), include_top = False, weights = "imagenet")
base_model.trainable = False
model = tf.keras.Sequential([base_model,
                                 tf.keras.layers.GlobalAveragePooling2D(),
                                 tf.keras.layers.Dropout(0.2),
                                 tf.keras.layers.Dense(4, activation="softmax")                                     
                                ])
base_learning_rate = 0.00001
model.compile(optimizer=tf.keras.optimizers.Adam(lr=base_learning_rate),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

history = model.fit(x_train,y_train,epochs = 125 , validation_data = (x_val, y_val))

```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(125)

plt.figure(figsize=(15, 15))
plt.subplot(2, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
